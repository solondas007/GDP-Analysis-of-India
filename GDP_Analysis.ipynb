```python
'''
For data preparation :
I set index as Items Description and removed the rows having 'GSDP - CURRENT PRICES (` in Crore)' as Items Description
because our analysis will be on the (% Growth over previous year) in year 2013-14, 2014-15 and 2015-16
'''
data_1a = data_1a.set_index("Items  Description")
data = data_1a. drop('GSDP - CURRENT PRICES (` in Crore)', axis = 0)

# the index was resetted again
data = data.reset_index()
```

```python
""" First part of the Question """
# Calculate the average growth of states for the duration 2013-14, 2014-15 and 2015-16 by taking the mean of the row '(% Growth over previous year)'

avg_growth_of_states = round(data.mean(),2)
avg_growth_of_state = avg_growth_of_states[0:-1]
```

```python
'''
Plotting the data using matplotlib

'''

# specifying the figure size
plt.figure(num=None, figsize=(15,15), dpi=80)
index = np.arange(len(df['States']))

# plotting a barplot
plt.bar(index, df['Average Growth of States'])
# naming and styling
plt.xlabel('States', fontsize=20)
plt.ylabel('Average Growth of States', fontsize=20)
plt.xticks(index, df['States'], fontsize=10, rotation=90)
plt.title('Average Growth of States for the duration 2013-14, 2014-15 and 2015-16',fontsize=25)

# displaying
plt.show()
```

#### 4) <br>
#### Categorise the states into four groups based on the GDP per capita (C1, C2, C3, C4, where C1 would have the highest per capita GDP and C4, the lowest). The quantile values are (0.20,0.5, 0.85, 1), i.e., the states lying between the 85th and the 100th percentile are in C1; those between the 50th and the 85th percentiles are in C2, and so on.

```python
'''
Filtering the original dataframe into a new one, sorting the dtatframe on the per capita GDP values
'''

df_group_states = df_total[['States','Per Capita GSDP (Rs.)']]
df_group_states = df_group_states.sort_values(by = 'Per Capita GSDP (Rs.)')

# Applying the quantile function to check in which quantile each group of ststes should belong 
df_group_states['Per Capita GSDP (Rs.)'].quantile([0.20,0.5, 0.85, 1])
```

```python
# Applying the qcut() to group the states according to the quantiles.
state_catagories = pd.qcut(df_group_states['Per Capita GSDP (Rs.)'],[0,0.20,0.5, 0.85,1],labels=["C4", "C3", "C2","C1"])

# Adding the catagories of state as a new column
df_group_states['Grouped State'] = state_catagories  
df_group_states
```

```python
'''
Filtering out the data and sorting based on Per Capita GSDP and adding our previously created state catagories..
'''


df_sub_sectors = df_total[['States','Agriculture, forestry and fishing','Mining and quarrying',
         'Manufacturing','Electricity, gas, water supply & other utility services','Construction',
         'Trade, repair, hotels and restaurants','Transport, storage, communication & services related to broadcasting',
         'Financial services','Real estate, ownership of dwelling & professional services','Public administration',
         'Other services','Gross State Domestic Product','Per Capita GSDP (Rs.)']]
df_sub_sectors = df_sub_sectors.sort_values(by  = 'Per Capita GSDP (Rs.)')
df_sub_sectors['State Catagory'] = state_catagories
```

```python
'''
Creating 4 dataframes for each state catagory C1,C2,C3,C4...

'''


df_c4 = df_sub_sectors.where(df_sub_sectors['State Catagory'] == 'C4')
df_c3 = df_sub_sectors.where(df_sub_sectors['State Catagory'] == 'C3')
df_c2 = df_sub_sectors.where(df_sub_sectors['State Catagory'] == 'C2')
df_c1 = df_sub_sectors.where(df_sub_sectors['State Catagory'] == 'C1')
```
